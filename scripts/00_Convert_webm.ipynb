```python
# ========================================================
# convert_webm_to_mp4.py
# ========================================================
import subprocess
from pathlib import Path

WEBM_FOLDER    = Path("/Users/cynthianyongesa/Desktop/Desktop - Cynthia's Macbook Pro/DATA/2_EYE_TRACKING_PY/Eye-Tracking Videos/WEBM")
OUTPUT_FOLDER  = Path("/Users/cynthianyongesa/Desktop/Desktop - Cynthia's Macbook Pro/DATA/2_EYE_TRACKING_PY/Eye-Tracking Videos/ORIGINAL_MP4")

OUTPUT_FOLDER.mkdir(parents=True, exist_ok=True)

webm_files = list(WEBM_FOLDER.glob("*.webm"))
print(f"Found {len(webm_files)} .webm files\n")

skipped   = []
converted = []
failed    = []

for webm_file in webm_files:
    mp4_file = OUTPUT_FOLDER / webm_file.with_suffix(".mp4").name

    # ── Skip if MP4 already exists ────────────────────────────────
    if mp4_file.exists():
        print(f"  ⏭  Skipping (already exists): {mp4_file.name}")
        skipped.append(mp4_file.name)
        continue

    # ── Convert ───────────────────────────────────────────────────
    print(f"  🔄 Converting: {webm_file.name} → {mp4_file.name}")
    cmd = [
        "ffmpeg", "-i", str(webm_file),
        "-c:v", "libx264",
        "-preset", "fast",
        "-crf", "22",
        "-c:a", "aac",
        "-y",           # overwrite if somehow it exists mid-run
        str(mp4_file)
    ]
    result = subprocess.run(cmd, capture_output=True, text=True)

    if result.returncode == 0:
        print(f"  ✅ Done: {mp4_file.name}")
        converted.append(mp4_file.name)
    else:
        print(f"  ❌ FAILED: {webm_file.name}")
        print(f"     {result.stderr[-300:]}")   # last 300 chars of error
        failed.append(webm_file.name)

# ── Summary ───────────────────────────────────────────────────────
print(f"\n{'='*50}")
print(f"  Converted:  {len(converted)}")
print(f"  Skipped:    {len(skipped)}  (MP4 already existed)")
print(f"  Failed:     {len(failed)}")
if failed:
    print(f"\n  Failed files:")
    for f in failed: print(f"    ✗ {f}")
print(f"{'='*50}")
```
